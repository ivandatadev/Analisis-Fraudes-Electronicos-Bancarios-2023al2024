## 2️⃣ Renombrado de Columnas

## 3️⃣ Indicadores Agregados

Se calculan indicadores clave como total de reclamos, montos involucrados, clientes afectados, y ratios de severidad.


```python
def calcular_indicadores(df):
    df['Total_Reclamos'] = df[['Reclamos_TC','Reclamos_TD','Reclamos_TPPF',
                               'Reclamos_Transf','Reclamos_Cajeros','Reclamos_Otros']].sum(axis=1)
    
    df['Total_Monto'] = df[['Monto_TC','Monto_TD','Monto_TPPF',
                            'Monto_Transf','Monto_Cajeros','Monto_Otros']].sum(axis=1)
    
    df['Total_Clientes'] = df[['Clientes_TC','Clientes_TD','Clientes_TPPF',
                               'Clientes_Transf','Clientes_Cajeros','Clientes_Otros']].sum(axis=1)
    
    df['Monto_promedio_por_reclamo'] = df['Total_Monto'] / df['Total_Reclamos']
    df['Reclamos_por_cliente'] = df['Total_Reclamos'] / df['Total_Clientes']
    return df
```
